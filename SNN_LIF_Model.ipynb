```python
import snntorch as snn
import torch
import matplotlib.pyplot as plt
```

```python
lif= snn.Leaky (beta=0.5)
```

```python
lif1= snn.Leaky (beta=0.95)
```

beta controls leak rate

```python
num_steps=100
membrane_potential = []
spikes =[]
```

```python
mem=torch.zeros(1)
```

```python
input_current =torch.zeros(num_steps)
input_current[20:80] = 0.5
```

```python
for step in range(num_steps):
    spk,mem=lif(input_current[step],mem)
    membrane_potential.append(mem.item())
    spikes.append(spk.item())
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(input_current)
plt.title('Input Current')
plt.subplot(3, 1, 2)
plt.plot(membrane_potential)
plt.title('Membrane Potential')
plt.axhline(y=1.0, color='r', linestyle='--', label='Threshold')
plt.subplot(3, 1, 3)
plt.plot(spikes)
plt.title('Output Spikes')
plt.tight_layout()
plt.show()
```
